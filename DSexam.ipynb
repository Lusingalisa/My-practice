```python
Diamonds.isnull().sum()
```

- We then plot histograms to find out if the data in depth is skewed or normally distributed

```python
plt.figure(figsize=(10,10))
for i, col in enumerate (Diamonds.columns):
    plt.hist(Diamonds["depth"])
    plt.title("Depth")
    plt.ylabel("frequency")
plt.tight_layout()
plt.show()
```

- We can see the data in depth is skewed

```python
Diamonds["depth"]=Diamonds["depth"].fillna(Diamonds["depth"].median())
```

```python
Diamonds.isnull().sum()
```

- we group the continuous data together as cont_data

```python
cont_data=Diamonds.select_dtypes(exclude="object")
cont_data
```

- We plot boxplots to see if our data has any outliers

```python
plt.figure(figsize=(10,10))
for i, col in enumerate (cont_data.columns):
    plt.subplot(3,3,i+1)
    plt.boxplot(cont_data[col])
    plt.title([col])
    plt.ylabel("frequency")
plt.tight_layout()
plt.show()
```

- The data set has outliers so we are going to define a function that removes outliers

```python
def Remove_outliers(column):
    for column in column:
        lower_quantile=cont_data[column].quantile(.25)
        upper_quantile=cont_data[column].quantile(.75)

        IQR=upper_quantile-lower_quantile

        upper_boundary=upper_quantile + 1.5 *IQR
        lower_boundary=lower_quantile -1.5 *IQR

        cont_data[column]=np.where(cont_data[column]>upper_boundary,upper_boundary,cont_data[column])
        cont_data[column]=np.where(cont_data[column]<lower_boundary,lower_boundary,cont_data[column])
```

```python
Remove_outliers(cont_data)
```

```python
plt.figure(figsize=(10,10))
for i, col in enumerate(cont_data.columns):
    plt.subplot(3,3,i+1)
    plt.boxplot(cont_data[col])
    plt.title(col)
    plt.ylabel("frequency")
plt.tight_layout()
plt.show()
```

- The boxplot shows that all outliers have been removed
